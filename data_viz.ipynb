<h2 style="color:#4CC9F0">Import Statements</h2>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h2 style="color:#4CC9F0">Viz Presentation</h2>

```python
sns.set(context="talk", style="darkgrid", palette="husl")
```

<h2 style="color:#4CC9F0">Data Sets</h2>

```python
iris_df = sns.load_dataset("iris")
iris_df.sample(2)
```

<p style="color:#CFBAF0">Scatter Plot</p>

```python
sns.relplot(data=iris_df, x="sepal_length", y="petal_length",
            hue="species", kind="scatter")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

<p style="color:#CFBAF0">Join Plot</p>

```python
sns.jointplot(data=iris_df, x="sepal_length", y="petal_length", hue="species",
              marginal_ticks="True")
plt.show()
```

<p style="color:#CFBAF0">Box Plot</p>

```python
plt.figure(dpi=110)
sns.boxplot(data=iris_df, x="species", y="sepal_length")
plt.show()
```

```python
sns.barplot(data=iris_df, x="species", y="petal_length")
plt.show()
```
